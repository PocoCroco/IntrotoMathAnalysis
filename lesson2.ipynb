```python
# № 2. Зададим множества A, B, C.
A = {46, 70, 85, 9, 4, 39}
B = {5, 70, 4, 10, 39, 75}
C = {43, 2, 46, 5, 85, 4, 10, 53, 24, 39}
```

```python
# Пересечение
A & B & C
```

```python
# Объединение
A | B | C
```

```python
# Разность
A - B - C
```

```python
#Симметрическая разность
A ^ B ^ C
```
